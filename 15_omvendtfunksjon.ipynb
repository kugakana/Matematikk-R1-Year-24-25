#### Omvendte funksjoner

La $ f $ være en entydig funksjon. En funksjon $ f^{-1}$ er den omvendte funksjonen til $ f $ hvis

$$ f^{-1}\left(f(x)\right) = x $$ 

<li> Definisjonsmengden til $f$ er verdimengden til $f^{-1}$ , og omvendt. </li>

<li> Grafen til $f$ og grafen til $f^{-1}$ ligger symmetrisk om linja $y = x$. </li>

<li> Vi tegner grafene til $f$ og $f^{-1}$ ved å bytte om koordinatene i verditabellen. </li>

<img width="350" height="350" src="bilde_10_omvendtefun.png">

#### Eksempel

Funksjonen $f$ er gitt  ved  $f(x) = x^3 + 2x \,\,\,\, , \,\,\ x \in [-2 , 2]$

Tegn grafen til $f$ og grafen til den omvendte funksjonen $f^{-1}$. 

```python
# importerer de nødvendige bibliotekene

import numpy as np                         # Kan skrive "from numpy import*"
import pylab as pl                         # Kan skrive "from numpy import*"

# Definer funksjonen
def f(x):
    return x**3 + 2*x

# Tegn grafen til funksjonen og den omvendte funksjonen

fig = pl.subplots(figsize=(8, 6))                         # Figurstørelse
x = np.linspace(-2 , 2 , 50)                              # Antall grid punkter er 50

pl.plot(x , f(x) , color = 'red' , label = 'f(x)')        # Tegn funksjonen
pl.plot(f(x) , x , color = 'blue',label = 'inverse f(x)') # Tegn omvendtfunksjon
x = np.linspace(-12 , 12 , 50)                       
pl.plot(x , x ,'--' , color = 'green' , label = 'y = x')  # Tegn linje y = x

pl.ylim(-12 , 12)                                         # Grenseverdier til y-aksen
pl.legend(loc='upper left')                               # Graf-etiketten

# Tegn x-aksen og y-aksen
pl.axhline(color = 'black')
pl.axvline(color = 'black')

# Gi navn til aksene og merk x og y akser
pl.xlabel('x')
pl.ylabel('y' , rotation = 0)
pl.show()
```

#### Oppgave

Funksjonen $f$ er gitt  ved  $f(x) = \sqrt(2x + 1) \,\,\,\, , \,\,\ x \in [-0.5 , 10]$

a) Skriv et program som lager verditabell over $f$ med steglengde i $x$ på 1,0.

b) Tegn grafen til $f$ og grafen til den omvendte funksjonen $f^{-1}$.

```python
# importerer de nødvendige bibliotekene

from numpy import *                        # Kan skrive "import numpy as np"
from pylab import *                        # Kan skrive "import pylab as py*"

# Definer funksjonen
def f(x):
    return sqrt(2*x + 1)

# Tegn grafen til funksjonen og den omvendte funksjonen og symmetri linjen
fig = subplots(figsize=(8, 8))                         # Figurstørelse
x = linspace(- 0.5 , 10 , 200)                         # Antall grid punkter er 200

plot(x , f(x) , color = 'red' , label = 'f(x)')        # Tegn funksjonen
plot(f(x) , x , color = 'blue',label = 'inverse f(x)') # Tegn omvendtfunksjon
plot(x , x ,'--' , color = 'green' , label = 'y = x')  # Tegn linje y = x

ylim(-2 , 10)                                          # Grenseverdier til y-aksen
legend(loc='upper right')                              # Graf-etiketten

# Tegn x-aksen og y-aksen
axhline(color = 'black')
axvline(color = 'black')

# Gi navn til aksene og merk x og y akser
xlabel('x')
ylabel('y' , rotation = 0)
show()
```
