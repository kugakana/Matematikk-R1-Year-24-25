### Vektorer

Lag et program i Python som tegner ut  av en vektor $ v âƒ—  =  [x ,y]  =  [3  ,4]$ 

```python
from matplotlib.pyplot import*
from numpy import*

# Vektor opprinnelse plassering 
X = [0] 
Y = [0] 

# Lag en posisjonsvektor
v = array([3,4])

# Lag plottet
fig, ax = subplots()

# Legg til vektoren v til plottet
ax.quiver(X, Y, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
title('Posisjonsvektor')

# Angi x-grenser og y-grenser for plottet
ax.set_xlim([-1, 5])
ax.set_ylim([-1, 5])

# Tegn x-aksen og y-aksen
axhline(color = 'black')
axvline(color = 'black')

# Vis plottet sammen med rutenettet
grid()
show()
```

#### Lengden av en vektor

Lag et program i Python som regner ut lengden av en vektor $u âƒ—  =  [x ,y]$.

Bruk programmet til Ã¥ finne lengden av vektoren $ u âƒ—  =  [3  ,4]$.


```python
from numpy import *

u = array([3,4])                                 # Skriv vektoren
lengde = sqrt(u[0]**2 + u[1]**2)                 # Lengden av vektor OA

# Skriv vektoren u og lengden av u i konsollen
print()
print(f'u = {u}')
print()
print(f'1. Lengden av vektoren u = {lengde:.2f}')
print()

# Bestem lengden av en vektor med kommando linlag.norm()
print(f'2. Lengden av vektoren u = {linalg.norm(u):.2f}')
```

```python
from matplotlib.pyplot import*
from numpy import*

# Vektor opprinnelse plassering 
X = [0] 
Y = [0] 

# Lag multippelvektorer
v = array([2 , 3])
u = array([5 , 1])
w = array([v[0] + u[0] , v[1] + u[1]])        # Summen av to vektorer

# Lag plottet
fig, ax = subplots()

# Legg til vektorene til plottet
ax.quiver(X, Y, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(X, Y, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.quiver(X, Y, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='g')
title('Addisjon av to vektorer')

# Angi x-grenser og y-grenser for plottet
ax.set_xlim([-1, 8])
ax.set_ylim([-1, 5])

# Tegn x-aksen og y-aksen
axhline(color = 'black')
axvline(color = 'black')

# Vis plottet sammen med rutenettet
grid()
show()
print(f'w = {w}')
```
